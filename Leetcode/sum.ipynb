```python
class Solution(object):
    def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        if len(s) != len(t):
            return False
        
        seens = {}
        seent ={}
        for idx in range(len(s)):
            if s[idx] not in seens and t[idx] not in seent:
                seens[s[idx]] = t[idx]
                seent[t[idx]] = s[idx]
            elif seens[s[idx]] != t[idx] or seent[t[idx]] != s[idx]:
                return False
            
        return True
    
    def twoSum(self, nums, target, l, r):
            """
            :type nums: List[int]
            :type target: int
            :rtype: List[int]
            """

            n = len(nums[l: r+1])
            if n < 2:
                return set()

            start = nums[l - 1]
            dict1 = dict()
            ret = []
            for i in range(l, r+1):
                other = target - nums[i]
                if other in dict1:
                    ret.append([dict1[other], nums[i]])
                dict1[nums[i]] = nums[i]

            return ret
        
    def t(self, nums, target):
        nums.sort()
        ret = []
        n = len(nums)
        for i in range(n-3):
            import pdb
#                 pdb.set_trace()
            for g in range(i + 1, n-2):
                tar = nums[i] + nums[g]

                if i > 0  and g > i + 1 and g < n-2 and  tar == nums[i] + nums[g - 1]:
                    continue

                t = self.twoSum(nums, target-tar, g + 1, n - 1)
                if not t:
                    continue
                list(map(lambda a: a.extend([nums[i], nums[g]]), t))
                t = filter(lambda a: not(a[0] == a[1] == a[2] == a[3]), t)
                ret.extend(t)

        return ret

    def romanToInt(self, s):
        """
        :type s: str
        :rtype: int
        """
        sv = {
            'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000
             }
        ret = 0
        s=list(reversed(s))
        for i, v in enumerate(s):
            if i > 0 and sv[v] < sv[s[i-1]]:
                ret -= sv[v]
                print(ret)
                continue
            ret += sv[v]
            
        return ret
    
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        flag = False
        ret = ''
        if not strs or not strs[0]:
            return ''
        
        for s in strs[0]:
            
            for st in strs:
                if not st.startswith(ret + s):
                    
                    return ret
            ret += s
        
        return ret
    
    def threeSumClosest(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        n = len(nums)
        if n < 3:
            return 
        nums.sort()
        cur = abs(sum(nums[:3]) - target)
        for i in range(n-2):
            l = i + 1
            r = n -1
            while l < r:
                
                s = nums[i] + nums[l] + nums[r]
                
                if s > target:
                    r -= 1
                elif s < target:
                    l += 1
                else:
                    return target
                
                if abs(cur-target) > abs(s - target):
                    cur = s
        
        return cur
                    
                
```
